# Analiziranje podatkov

```python
# naložimo paket
import os
import pandas as pd

nov_directory = 'obdelani podatki'
csv_filename = 'vsi_podatki.csv'

# izberemo interaktivni "notebook" stil risanja
%matplotlib inline

pd.set_option("display.max_rows", 15)
path = os.path.join(nov_directory, csv_filename) 

#naložimo razpredelnico, s katero bomo delali
countries = pd.read_csv(path)
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
countries.sort_values(by="visina moskih", ascending = False, inplace = True)
countries
print(countries.head(5))
countries["visina moskih"]
```

```python
countries.sort_values(by="visina zensk", ascending = False, inplace = True)
countries
print(countries.head(5))
countries["visina zensk"]
```

```python
countries.sort_values(by="visina moskih", ascending = True, inplace = True)
countries
print(countries.head(5))
countries["visina moskih"]
```

```python
countries.sort_values(by="visina zensk", ascending = True, inplace = True)
countries
print(countries.head(5))
countries["visina zensk"]
```

## 2. Podtabela in grafi

```python
podtabela = countries[["ime", "visina moskih", "teza moskih", "bmi moskih", "visina zensk", "teza zensk", "bmi zensk"]]
podtabela
```

```python
#predstavlja drzave glede na različna povprečja
podtabela.hist()
```

```python
countries_by_maleheight = countries.groupby('visina moskih')
```

```python
countries_by_maleheight.size().plot.bar(x='ime', y='visina moskih', color='C1', title='Visina moskih po državah', figsize=(8, 4))
```

# Analiziranje podatkov

```python
# naložimo paket
import os
import pandas as pd

nov_directory = 'obdelani podatki'
csv_filename = 'vsi_podatki.csv'

# izberemo interaktivni "notebook" stil risanja
%matplotlib inline

pd.set_option("display.max_rows", 15)
path = os.path.join(nov_directory, csv_filename) 

#naložimo razpredelnico, s katero bomo delali
countries = pd.read_csv(path)
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
countries.sort_values(by="visina zensk", ascending = False, inplace = True)
countries
print(countries.head(5))
countries["visina zensk"]
```

```python
countries.sort_values(by="visina moskih", ascending = True, inplace = True)
countries
print(countries.head(5))
countries["visina moskih"]
```

```python
countries.sort_values(by="visina zensk", ascending = True, inplace = True)
countries
print(countries.head(5))
countries["visina zensk"]
```

## 2. Podtabela

```python
podtabela = countries[["ime drzave", "visina moskih", "teza moskih", "bmi moskih", "visina zensk", "teza zensk", "bmi zensk"]]
podtabela
```

## 3. Grafi

```python
#predstavlja drzave glede na različna povprečja
podtabela.hist()
```

```python

countries_by_maleheight = countries.groupby('ime drzave')['visina moskih'].mean() #mean izračuna povprečje od vsakega stolpca

plt.figure(figsize=(80, 8))
countries_by_maleheight.plot(kind='bar', color='C0')
plt.xlabel('Ime države')
plt.ylabel('Višina moških')
plt.title('Višina moških po državah')
plt.xticks(rotation=55)  # rotacija, da se lažje prebere vse države
plt.tight_layout()  # za urejeno razporeditev imen
plt.show()


countries_by_femaleheight = countries.groupby('ime drzave')['visina zensk'].mean() #mean izračuna povprečje od vsakega stolpca

plt.figure(figsize=(80, 8))
countries_by_femaleheight.plot(kind='bar', color='C1')
plt.xlabel('Ime države')
plt.ylabel('Višina žensk')
plt.title('Višina žensk po državah')
plt.xticks(rotation=55)  # rotacija, da se lažje prebere vse države
plt.tight_layout()  # za urejeno razporeditev imen
plt.show()

```

### Razlike višin med spoloma

Povprečna višina moškega in ženske glede na vse države

```python
average_world_male_height = round(np.mean(podtabela["visina moskih"]), 2)
average_world_female_height = round(np.mean(podtabela["visina zensk"]), 2)

print("Povprečna višina moškega glede na vse države je", average_world_male_height, "m.")
print("Povprečna višina ženske glede na vse države je", average_world_female_height, "m.")
```

Koliko so moški večji od žensk?

```python
rounded_height_difference = round(average_world_male_height - average_world_female_height, 2)

print("Razlika med višino moških in žensk je", rounded_height_difference, "m.")
```
